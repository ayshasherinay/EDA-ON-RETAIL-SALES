```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\AYSHA SHERIN\retail_sales_dataset.csv')
```

```python
print(df.head())
```

```python
# Convert 'Date' column to datetime if necessary
df['Date'] = pd.to_datetime(df['Date'])

# Handle missing values
df.dropna(inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Get a summary of the data
print(df.describe())

# View the first few rows of the cleaned data
print(df.head())
```

```python
mean_age = df['Age'].mean()
```

```python
mean_quantity = df['Quantity'].mean()
mean_price = df['Price per Unit'].mean()
mean_total_amount = df['Total Amount'].mean()

print(f"Mean Age: {mean_age}")
print(f"Mean Quantity: {mean_quantity}")
print(f"Mean Price per Unit: {mean_price}")
print(f"Mean Total Amount: {mean_total_amount}")
```

```python
median_age = df['Age'].median()
median_quantity = df['Quantity'].median()
median_price = df['Price per Unit'].median()
median_total_amount = df['Total Amount'].median()

print(f"Median Age: {median_age}")
print(f"Median Quantity: {median_quantity}")
print(f"Median Price per Unit: {median_price}")
print(f"Median Total Amount: {median_total_amount}")
```

```python
mode_age = df['Age'].mode()[0]
mode_quantity = df['Quantity'].mode()[0]
mode_price = df['Price per Unit'].mode()[0]
mode_total_amount = df['Total Amount'].mode()[0]

print(f"Mode Age: {mode_age}")
print(f"Mode Quantity: {mode_quantity}")
print(f"Mode Price per Unit: {mode_price}")
print(f"Mode Total Amount: {mode_total_amount}")
```

```python
# Calculate variance and standard deviation
variance_age = df['Age'].var()
std_dev_age = df['Age'].std()

variance_quantity = df['Quantity'].var()
std_dev_quantity = df['Quantity'].std()

variance_price = df['Price per Unit'].var()
std_dev_price = df['Price per Unit'].std()

variance_total_amount = df['Total Amount'].var()
std_dev_total_amount = df['Total Amount'].std()

# Summary statistics for all numeric columns
summary_stats = df.describe()

# Print results
print(f"Mean Age: {mean_age}, Median Age: {median_age}, Mode Age: {mode_age}")
print(f"Mean Quantity: {mean_quantity}, Median Quantity: {median_quantity}, Mode Quantity: {mode_quantity}")
print(f"Mean Price per Unit: {mean_price}, Median Price per Unit: {median_price}, Mode Price per Unit: {mode_price}")
print(f"Mean Total Amount: {mean_total_amount}, Median Total Amount: {median_total_amount}, Mode Total Amount: {mode_total_amount}")

print(f"Variance Age: {variance_age}, Std Dev Age: {std_dev_age}")
print(f"Variance Quantity: {variance_quantity}, Std Dev Quantity: {std_dev_quantity}")
print(f"Variance Price per Unit: {variance_price}, Std Dev Price per Unit: {std_dev_price}")
print(f"Variance Total Amount: {variance_total_amount}, Std Dev Total Amount: {std_dev_total_amount}")

# Print summary statistics
print(summary_stats)
```

```python
df = pd.read_csv(r'C:\Users\AYSHA SHERIN\retail_sales_dataset.csv', parse_dates=['Date'])
```

```python
print(df.columns)
```

```python
# Set 'Date' as the index for resampling
df.set_index('Date', inplace=True)

yearly_sales = df['Total Amount'].resample('Y').sum()


# View the resampled data
print(yearly_sales.head())
```

```python
import matplotlib.pyplot as plt

# Plot monthly sales
monthly_sales.plot(kind='line', figsize=(10,6), title='Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Group the data by month to check for seasonality
df['Month'] = df.index.month
monthly_seasonality = df.groupby('Month')['Total Amount'].sum()

# Plot the seasonality
monthly_seasonality.plot(kind='bar', figsize=(10,6), title='Sales by Month (Seasonality)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Calculate the 12-month rolling average of sales
rolling_sales = monthly_sales.rolling(window=12).mean()

# Plot the original monthly sales and the rolling average
plt.figure(figsize=(10,6))
monthly_sales.plot(label='Monthly Sales')
rolling_sales.plot(label='12-Month Rolling Average', color='red')
plt.title('Monthly Sales with Rolling Average')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Group sales by 'Age' and calculate the mean sales for each age group
df.groupby('Age')['Total Amount'].mean().plot(kind='bar', figsize=(10,6), title='Average Sales by Age Group')
plt.xlabel('Age')
plt.ylabel('Average Sales')
plt.show()
```

```python
# Group by gender and calculate total sales for each gender
gender_sales = df.groupby('Gender')['Total Amount'].sum()

# Plot total sales by gender
gender_sales.plot(kind='bar', figsize=(10,6), title='Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Group by 'Product Category' and sum the 'Total Amount' to get the top-selling products
top_products = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False).head(5)

# Display the top 5 products by sales
print(top_products)
```

```python
# Least-selling products (bottom 5)
least_products = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=True).head(5)

# Display the least 5 products by sales
print(least_products)
```

```python
# Calculate the total sales for each customer
customer_lifetime_value = df.groupby('Customer ID')['Total Amount'].sum()

# Display the top 5 customers by CLV
print(customer_lifetime_value.sort_values(ascending=False).head())
```

```python
# Calculate the average sales per customer
average_sales_per_customer = df.groupby('Customer ID')['Total Amount'].mean()

# Display the top 5 customers by average purchase value
print(average_sales_per_customer.sort_values(ascending=False).head())
```

```python
# Calculate the number of transactions per customer
purchase_frequency = df.groupby('Customer ID')['Transaction ID'].count()

# Display customers with the highest purchase frequency
print(purchase_frequency.sort_values(ascending=False).head())
```

```python
# Create a DataFrame for CLV with total sales, average sales, and purchase frequency
clv_df = df.groupby('Customer ID').agg({
    'Total Amount': ['sum', 'mean'],
    'Transaction ID': 'count'
}).rename(columns={'sum': 'Total Sales', 'mean': 'Avg Sales', 'count': 'Purchase Frequency'})

# Flatten the MultiIndex columns
clv_df.columns = clv_df.columns.droplevel(0)

# Display the top customers by total sales
print(clv_df.sort_values('Total Sales', ascending=False).head())
```

```python
# Count the number of occurrences of each product category
df['Product Category'].value_counts().plot(kind='bar', figsize=(10,6), title='Product Category Distribution')
plt.xlabel('Product Category')
plt.ylabel('Number of Sales')
plt.show()
```

```python
# Total sales by gender
df.groupby('Gender')['Total Amount'].sum().plot(kind='bar', figsize=(10,6), title='Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure 'Date' is set as the index and in datetime format
df.index = pd.to_datetime(df.index)

# Sort the DataFrame by date
df = df.sort_index()

# Aggregate sales by date
daily_sales = df['Total Amount'].resample('D').sum()

# Plot the aggregated sales data
daily_sales.plot(kind='line', figsize=(10,6), title='Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# 12-month rolling average for smoothing short-term fluctuations
df['Total Amount'].rolling(window=12).mean().plot(kind='line', figsize=(10,6), title='12-Month Rolling Average of Sales')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix for numerical columns
correlation_matrix = numeric_df.corr()

# Create a heatmap with annotations
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plot histogram of 'Total Amount' with 20 bins
df['Total Amount'].plot(kind='hist', bins=20, figsize=(10,6), title='Sales Distribution')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram for Age distribution
df['Age'].plot(kind='hist', bins=10, figsize=(10,6), title='Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Total Amount Histogram
df['Total Amount'].plot(kind='hist', bins=20, ax=axs[0, 0], title='Sales Distribution')

# Age Histogram
df['Age'].plot(kind='hist', bins=10, ax=axs[0, 1], title='Customer Age Distribution')

# Gender-based sales
df.groupby('Gender')['Total Amount'].sum().plot(kind='bar', ax=axs[1, 0], title='Total Sales by Gender')

# Sales trend over time
df['Total Amount'].plot(kind='line', ax=axs[1, 1], title='Sales Trend Over Time')

plt.tight_layout()
plt.show()
```
